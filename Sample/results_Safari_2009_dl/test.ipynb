```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os,sys
plt.rcParams['mathtext.fontset'] = 'stix'
plt.rcParams['font.family'] = 'STIXGeneral'
plt.rcParams['lines.linewidth'] = 2
plt.rcParams['lines.markersize'] = 2
plt.rcParams['xtick.major.pad'] = '8'  # set the gap between the labels and the axis
plt.rcParams['ytick.major.pad'] = '5'
plt.rcParams.update({'figure.autolayout': True})
plt.rc('axes', linewidth=1.8)
plt.rc('axes', linewidth=1.8)
```

```python
cap = pd.read_csv('condensated.txt', sep='\t', header=None)
cap
```

```python
cap.columns = ['t', 'capacity','non']
```

```python
cap = cap[['t','capacity']]
cap.columns = ['time','Capacity fading']
```

```python
cap.plot('time','Capacity fading')
#makr ticks bigger
plt.xticks(fontsize=15) 
plt.yticks(fontsize=15)
#make labels bigger
plt.xlabel('Time (s)',fontsize=20)
plt.ylabel('Capacity',fontsize=20)
# make a fit line over average capacity
fit = np.polyfit(cap['time'],cap['Capacity fading'],1)
#plot fit line
plt.plot(cap['time'],fit[0]*cap['time']+fit[1],color='red')


# #make legend
plt.legend(['Capacity','Average capacity'],fontsize=15)
plt.savefig('capacity.png',dpi=300)
```
